7. Todo el proceso realizado hasta acá, corresponde a la **Extracción de Datos (ETL)** y damos comienzo al proceso de **Transformación**.

Al realizar un análisis previo a la "Transformación", cuando implementamos la visualización de la información de columnas y tipos de datos "unificado_df.info()" notamos que en la columna "age" el número de datos era menor a la cantidad de filas que contiene el DataFrame. En este caso teníamos 2 opciones, recurrir a la eliminación de filas con datos vacíos (que podían afectar a un análisis posterior) o rellenar los valores vacíos/nulos con el promedio, mediana o un valor fijo a través del método ".fillna". Al observar que la diferencia de valores no era tan grande (119 valores de 99.457), optamos eliminar las filas sin datos.

```python
unificado_df.dropna(inplace=True)
unificado_df
```

**Conclusión:** Como pudimos observar al solicitar los datos de las columnas (unificado_df.info()) la columna "age" tenía una menor cantidad de datos que el resto, entonces, al aplicar una limpieza del DataFrame eliminando los valores nulos, nos encontramos que se han eliminado 119 filas, ya que se redujeron de 99.457 a 99.338 filas.

8. Se transforman los datos del formato dd/mm/YYYY al formato YYYY/mm en una nueva columna para obtener un mejor resultado para análisis posteriores.

```python
unificado_df['invoice_date'] = pd.to_datetime(unificado_df['invoice_date'], format='%d-%m-%Y')
unificado_df['year_month'] = unificado_df['invoice_date'].dt.to_period('M').astype(str)
fecha_global_df = unificado_df['year_month']
unificado_df
```

**Conclusión:** Se observa con un análisis superficial al visualizar los datos que las fechas (año/mes) coinciden con las originales. A su vez, observamos el incremento en la cantidad de columnas (11).

9. Se procede a borrar la columna “invoice_date” que contenía las fechas originales de los archivos .csv.

```python
unificado_df.drop('invoice_date', axis=1, inplace=True)
unificado_df
```

**Conclusión:** Se elimina la columna "invoice_date" y se visualizan los datos para corroborar.

10. Generamos una nueva columna “rango_etario” que contenga la siguiente clasificación: (>24), (25 a 35), (36 a 45), (46 a 60), (<61).

```python
bins = [0, 24, 35, 45, 60, float('inf')]
labels = ['>24', '25 a 35', '36 a 45', '46 a 60', '<61']
unificado_df['rango_etario'] = pd.cut(unificado_df['age'], bins=bins, labels=labels)
unificado_df
```

**Conclusión:** Para generar una nueva columna (rango_etario) para clasificar a los consumidores acorde a su edad, se implementa la función de Pandas pd.cut(). Primero se definen los límites (bins) y se nombran los rangos (labels). Al finalizar, solicitamos visualizar el DataFrame para un chequeo general. Corroboramos a simple vista que coinciden las edad con los rangos asignados y que l DataFrame se incrementa en una columna más (11).

11. Para dejar más limpio y preparado el DataFrame para el proceso de Carga, se eliminan las columnas “invoice_no” y la columna “quantity”, ya que no se utilizarán para los análisis posteriores.

```python
unificado_df.drop(['invoice_no', 'quantity'], axis=1, inplace=True)
unificado_df
```

**Conclusión:** Al borrar las 2 columnas mencionadas y solicitar visualizar el DataFrame, observamos que la cantidad de columnas se reduce a 9.

12. Para pasar a la etapa de **Carga (LOAD)**, generamos un nuevo DataFrame limpio, es decir, con todas las modificaciones/actualizaciones realizadas en la etapa de Transformación.

```python
nuevo_df = unificado_df.copy()
nuevo_df.info()
```

**Conclusión:** Generamos un nuevo DataFrame realizando una copia de unificado_df y mostramos la información de las columnas.

13. Aplicamos restricciones de integridad para asegurar la validez de los datos.

```python
initial_count = nuevo_df.shape[0]
nuevo_df = nuevo_df[(nuevo_df['age'] >= 17) & (nuevo_df['age'] <= 80)]
final_count = nuevo_df.shape[0]
print (f"Filas iniciales: {initial_count}, Filas después del filtro: {final_count}")
```

**Conclusión:** En este caso, aplicamos una restricción a la columna "edad", donde solicitamos un conteo inicial de filas, luego aplicamos un filtro que guarde las edades entre 18 y 80 años, y que realice un conteo final. Al imprimir los resultados, corroboramos que tiene la misma cantidad de filas.

**Proceso de ANALISIS DE DATOS.**

A. Cantidades y Porcentaje de Mujeres y Hombres que contiene el DataFrame.

```python
porcentaje_genero = (nuevo_df['gender'].value_counts(normalize=True) * 100).round(2)
porcentaje = pd.DataFrame({'Porcentaje (%)': porcentaje_genero})
resultadoA_df = pd.merge(nuevo_df, porcentaje, on='gender', how='left')
resultadoA_df = resultadoA_df.groupby(['gender', 'Porcentaje (%)']).size().reset_index(name='count')
resultadoA_df
```

**Conclusión:**
- línea1: realiza un conteo de la columna "gender" y lo transformamos a porcentaje con 2 decimales.
- línea2: transformamos los porcentajes en un dataframe
- línea3: combinamos los DataFrames con referencia a la columna "gender". (how='left') asegura que se mantengan los datos de nuevo_df y que los porcentajes se añadan a cada fila correspondiente.
- línea4: agrupa y convierte el DataFrame en columnas: "gender", Porcentaje" y "count".

B. Porcentaje de mujeres y hombres por rango etario

```python
genero_rango_df = nuevo_df.groupby(['gender', 'rango_etario']).size().reset_index(name='Cantidad')
genero_rango_df
```

C. Cantidad de consumidores por edad

```python
genero_edad_df = nuevo_df.groupby(['gender', 'age']).size().reset_index(name='Cantidad')
genero_edad_df
```

D. Modo de pago más frecuente de todos los clientes categorizados por su género.

```python
pago_genero_df = nuevo_df.groupby(['gender', 'payment_method']).size().reset_index(name='Cantidad')
pago_genero_df = pago_genero_df.sort_values(by='Cantidad', ascending=False).reset_index(drop=True)
pago_genero_df
```

**Conclusión:** Podemos observar los datos organizados por género (femenino y masculino) organizados desde el método de pago más utilizado al menos utilizado.

E. Queremos determinar el método de pago más utilizado por el rango etario.

```python
pago_rango_df = nuevo_df.groupby(['payment_method', 'rango_etario']).size().reset_index(name='Cantidad')
pago_rango_df = pago_rango_df.sort_values(by='Cantidad', ascending=False).reset_index(drop=True)
pago_rango_df
```

F. Queremos determinar los precios de los productos por categoría

```python
precio_cat_df = nuevo_df.groupby(['category', 'price']).size().reset_index(name='Cantidad')
precio_cat_df
```

G. Solicitamos que nos muestre el precio mínimo, máximo y un promedio por categoría.

```python
precios_por_categoria_df = nuevo_df.groupby('category')['price'].agg(['min', 'max', 'mean']).reset_index()
precios_por_categoria_df.rename(columns={'min': 'precio_minimo', 'max': 'precio_maximo', 'mean': 'precio_promedio'}, inplace=True)
precios_por_categoria_df
```

H. Solicitamos visualizar las categorías de los productos comprados por género y método de pago.

```python
categoria_genero_df = nuevo_df.groupby(['category', 'gender', 'payment_method']).size().reset_index(name='Cantidad')
categoria_genero_df = categoria_genero_df.sort_values(by='Cantidad', ascending=False).reset_index(drop=True)
categoria_genero_df
```

N. Solicitamos visualizar las categorías de compra acorde a la fecha (año/mes).

```python
nuevo_df['year_month'] = pd.to_datetime(nuevo_df['year_month'], format='%Y-%m')
nuevo_df['Year'] = nuevo_df['year_month'].dt.year
nuevo_df['Month'] = nuevo_df['year_month'].dt.month
categoria_month_year_df = nuevo_df.groupby(['Month', 'Year', 'category']).size().reset_index(name='Cantidad')
categoria_month_year_df = categoria_month_year_df.sort_values(by=['Month', 'Year']).reset_index(drop=True)
categoria_month_year_df
```
